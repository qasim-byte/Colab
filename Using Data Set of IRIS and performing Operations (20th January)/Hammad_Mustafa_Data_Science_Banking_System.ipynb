```python
#Using Pandas to read the iris dataset
import pandas as pd
df = pd.read_csv(r"C:\Muhammad Qasim Ashfaq\University 3A\DATA SCIENCE\Using Data Set of IRIS and performing Operations (20th January)\iris.data", header = None)
display(df.head(100))   #Disoplaying first 100 rows of the dataframe
```

```python
#Renaming the columns of the dataframe
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
display(df.head(100))   #Disoplaying first 100 rows of the dataframe with new column names
```

```python
# Checking for missing values in the dataframe
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
display(missing_values[missing_values>0])
```

```python
#using bfill and ffill
obj = df.copy() #create a copy of the dataframe
obj = obj.bfill()   #backward fill
obj = obj.ffill()   #forward fill
df_filled = obj #assigning the filled dataframe to a new variable
display(df_filled.head(100))        #Displaying first 100 rows of the filled dataframe
```

```python
#arranging the dataframe in ascending order based on sepal_length
df_sorted = df.sort_values(by='sepal_length', ascending=True)
display(df_sorted.head(100))   #Displaying first 100 rows of the sorted dataframe
```

```python
#array slicing
df_slice = df.iloc[10:21, 1:4]  #slicing rows from index 10 to 20 and columns from index 1 to 3
display(df_slice)   #Displaying the sliced dataframe
```

```python
#tuple to dataframe
data_tuple = [(5.1, 3.5, 1.4, 0.2, 'Iris-setosa'),
              (4.9, 3.0, 1.4, 0.2, 'Iris-setosa'),
              (4.7, 3.2, 1.3, 0.2, 'Iris-setosa')]
df_from_tuple = pd.DataFrame(data_tuple, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
display(df_from_tuple)   #Displaying the dataframe created from tuple
```

```python
#list to dataframe
data_list = [[5.1, 3.5, 1.4, 0.2, 'Iris-setosa'],
              [4.9, 3.0, 1.4, 0.2, 'Iris-setosa'],
              [4.7, 3.2, 1.3, 0.2, 'Iris-setosa']]
df_from_list = pd.DataFrame(data_list, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
display(df_from_list)   #Displaying the dataframe created from list
```

```python
#dictionary to dataframe
data_dict = {'sepal_length': [5.1, 4.9, 4.7],
              'sepal_width': [3.5, 3.0, 3.2],
              'petal_length': [1.4, 1.4, 1.3],
              'petal_width': [0.2, 0.2, 0.2],
              'class': ['Iris-setosa', 'Iris-setosa', 'Iris-setosa']}
df_from_dict = pd.DataFrame(data_dict)
display(df_from_dict)   #Displaying the dataframe created from dictionary
```

```python
#setting index of the dataframe to 'class' column
df_indexed = df.set_index('class')
display(df_indexed.head(100))   #Displaying first 100 rows of the dataframe with 'class' as index
```

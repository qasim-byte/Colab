```python
path = "C:\\Muhammad Qasim Ashfaq\\University 3A\\Data Science\\27th January\\kidney_disease_dataset.csv"
import pandas as pd
data = pd.read_csv(path) # Load the dataset
print(data.head(20)) # Display the first 20 rows of the dataset
```

```python
# Renaming columns for better readability
data.rename(columns={"id": "patient_id"}, inplace=True)
data.rename(columns={"bp": "blood_pressure"}, inplace=True)
data.rename(columns={"sg": "specific_gravity"}, inplace=True)
data.rename(columns={"al": "albumin"}, inplace=True)
data.rename(columns={"su": "sugar"}, inplace=True)
data.rename(columns={"rbc": "red_blood_cells"}, inplace=True)
data.rename(columns={"pc": "pus_cell"}, inplace=True)
data.rename(columns={"pcc": "pus_cell_clumps"}, inplace=True)
data.rename(columns={"ba": "bacteria"}, inplace=True)
data.rename(columns={"bgr": "blood_glucose_random"}, inplace=True)
data.rename(columns={"bu": "blood_urea"}, inplace=True)
data.rename(columns={"sc": "serum_creatinine"}, inplace=True)
data.rename(columns={"sod": "sodium"}, inplace=True)
data.rename(columns={"pot": "potassium"}, inplace=True)
data.rename(columns={"hemo": "hemoglobin"}, inplace=True)
data.rename(columns={"pcv": "packed_cell_volume"}, inplace=True)
data.rename(columns={"wc": "white_blood_cell_count"}, inplace=True)
data.rename(columns={"rc": "red_blood_cell_count"}, inplace=True)
data.rename(columns={"htn": "hypertension"}, inplace=True)
data.rename(columns={"dm": "diabetes_mellitus"}, inplace=True)
data.rename(columns={"cad": "coronary_artery_disease"}, inplace=True)
data.rename(columns={"appet": "appetite"}, inplace=True)
data.rename(columns={"pe": "pedal_edema"}, inplace=True)
data.rename(columns={"ane": "anemia"}, inplace=True)
print(data.head(20))
```

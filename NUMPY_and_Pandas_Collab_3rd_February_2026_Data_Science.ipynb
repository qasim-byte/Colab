```
import pandas as pd
import numpy as np

wine_df = pd.read_csv('/content/winequality-red.csv', sep=';')
display(wine_df.head())
np.mean(wine_df,axis = 0)
```

```
#What exactly is axis and what effect does it have on the displayed content
# In NumPy and Pandas, the 'axis' parameter specifies the dimension along which an operation is performed.
# - When 'axis=0' (the default for some functions), the operation is applied column-wise. This means the function will compute a result for each column.
#   For example, `np.mean(df, axis=0)` calculates the mean of each column in the DataFrame `df`.
# - When 'axis=1', the operation is applied row-wise. This means the function will compute a result for each row.
# In the executed cells, `np.mean(wine_df, axis=0)` and `np.median(wine_df, axis=0)` calculate the mean and median for each column, respectively, across all rows of the `wine_df` DataFrame.
```

```
np.median(wine_df,axis = 0)
np.median(wine_df) #here no axis is used so the code will flatten the DataFrame and calculate a single median value from all the data.
```

```
print("Percentile at 50% :",np.percentile(wine_df,50))
print()
print("Percentile at 75% :",np.percentile(wine_df,75))
print()
print("Percentile at 25% :",np.percentile(wine_df,25))
print()
print("Percentile at 100% :",np.percentile(wine_df,100))
print()
print("Percentile at 0% :",np.percentile(wine_df,0))

print("\n\tNote that the percentile at 50% and the Median of the Data Frame is same (3.12)")
```

```
print("The np.average(wine_df) is :",np.average(wine_df))
print("This Average value is the same as the Mean")
print("The Mean is :",np.mean(wine_df))
```

Standard deviation (SD) is a statistical measure of how dispersed or spread out data values are relative to their mean (average). A low SD indicates data points are clustered closely around the mean, while a high SD shows data is spread over a wider range. It is calculated as the square root of the variance.

```
np.std(wine_df) # Standard Deviation
```

Variance is a statistical measure of the dispersion or spread of data points around their mean (average) value . A high variance indicates widely scattered data, while low variance indicates data clustered close to the mean. It is calculated as the average of the squared differences from the mean, and its square root is the standard deviation.

```
np.var(wine_df) #variance
```
